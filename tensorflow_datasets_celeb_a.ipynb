## Import TensorFlow, Keras, and TensorFlow datasets

````
# Import tensorflow and keras
%tensorflow_version 2.x
import tensorflow as tf
from tensorflow import keras

# Import tensorflow datasets
import tensorflow_datasets as tfds
````

## Print names of all datasets

````
# Print all tensorflow dataset
print(tfds.list_builders())
````

## Load the celebrities dataset ```celeb_a```

You can find more info about about this dataset at [celeb_a](https://www.tensorflow.org/datasets/catalog/celeb_a).  Note that it takes a while to download the dataset for the first time.



````
# Construct a tf.data.Dataset
dataset = tfds.load(name='celeb_a', split='train')

# Build your input pipeline
dataset = dataset.shuffle(1024).batch(16).prefetch(tf.data.experimental.AUTOTUNE)
for features in dataset.take(1):
  image, attributes = features['image'], features['attributes']
````

## Display some celebrities

````
%matplotlib inline
import matplotlib.pyplot as plt

def show_batch(image, attributes):
    plt.figure(figsize=(20, 20))
    for n in range(16):
        ax = plt.subplot(4, 4, n+1)
        plt.imshow(image[n].numpy())
        # male/female
        if (attributes['Male'][n]).numpy():
            title = 'male'
        else:
            title = 'female'
        # smiling/not smiling
        if (attributes['Smiling'][n]).numpy():
            title += ' smiling'
        else:
            title += ' not smiling'
        # set title
        plt.title(title)
        plt.axis('off')
````

````
print(image.shape)
````

````
show_batch(image, attributes)
````
